```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn import tree
import matplotlib.pyplot as plt
from IPython.display import display, Image
import ipywidgets as widgets
import json
```

```python
import ipywidgets as widgets
from IPython.display import display

# create text box
text = widgets.Text(
    placeholder='JSON string',
    description='Input:',
)

# create button
button = widgets.Button(description='Submit')

# display widget
display(text, button)

# define button click event
def on_button_click(b):
    global data_dict
    try:
        data_dict = json.loads(text.value)  # parse input string to Python dictionary
        print(f'Input parsed and stored in variable "user_input": {data_dict}')
    except json.JSONDecodeError:
        print("Invalid JSON input. Please provide a valid JSON string.")


button.on_click(on_button_click)

# example input
# {
#     "Alt": ["Yes", "Yes", "No", "Yes", "Yes", "No", "No", "No", "No", "Yes", "No"],
#     "Bar": ["No", "No", "Yes", "No", "No", "Yes", "Yes", "No", "Yes", "Yes", "No"],
#     "Fri": ["No", "No", "No", "Yes", "Yes", "No", "No", "No", "Yes", "Yes", "No"],
#     "Hun": ["Yes", "Yes", "No", "Yes", "No", "Yes", "No", "Yes", "No", "Yes", "No"],
#     "Pat": ["Some", "Full", "Some", "Full", "Full", "Some", "None", "Some", "Full", "Full", "None"],
#     "Price": ["$$$", "$", "$", "$", "$$$", "$$", "$", "$$", "$", "$$$", "$"],
#     "Rain": ["No", "No", "No", "Yes", "No", "Yes", "Yes", "Yes", "Yes", "No", "No"],
#     "Res": ["Yes", "No", "No", "No", "Yes", "Yes", "No", "Yes", "No", "Yes", "No"],
#     "Type": ["French", "Thai", "Burger", "Thai", "French", "Italian", "Burger", "Thai", "Burger", "Italian", "Thai"],
#     "Est": ["0-10", "30-60", "0-10", "10-30", ">60", "0-10", "0-10", "0-10", ">60", "10-30", "0-10"],
#     "WillWait": ["Yes", "No", "Yes", "Yes", "No", "Yes", "No", "Yes", "No", "No", "No"]
# }
```

```python
# Load data into a pandas DataFrame
df = pd.DataFrame(data_dict)

# Preprocess the dataset
le = LabelEncoder()

for column in df.columns:
    df[column] = le.fit_transform(df[column])
    
df
```

```python
# Split the dataset into training and test set
X = df.drop('Type', axis=1)
y = df['Type']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)  # 70% training and 30% test
```

```python
# Train the decision tree
clf = DecisionTreeClassifier(criterion="entropy")
clf = clf.fit(X_train,y_train)
```

```python
# Predict the test set results
y_pred = clf.predict(X_test)

y_pred
```

```python
# Print out the accuracy
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```

```python
# Plot the tree
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=200)

tree.plot_tree(clf, 
               feature_names = X.columns, 
               class_names=['Burger', 'French', 'Italian', 'Thai'],
               filled = True);

# output
```
